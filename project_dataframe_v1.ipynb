### Creating dataset with historical financial indicators for machine learning project 

```python

###
%matplotlib inline
import pandas as pd
import numpy as np
```

```python
# setting datasets path
ttm = './datasets/'
```

```python
# first step, importing csv files with financial information from Simfin
ttm_18 = pd.read_csv(ttm + 'ttm_09_30_2018.csv')
ttm_17 = pd.read_csv(ttm + 'ttm_09_30_2017.csv')
ttm_16 = pd.read_csv(ttm + 'ttm_09_30_2016.csv')
ttm_15 = pd.read_csv(ttm + 'ttm_09_30_2015.csv')
ttm_14 = pd.read_csv(ttm + 'ttm_09_30_2014.csv')
ttm_13 = pd.read_csv(ttm + 'ttm_09_30_2013.csv')
ttm_12 = pd.read_csv(ttm + 'ttm_09_30_2012.csv')
ttm_11 = pd.read_csv(ttm + 'ttm_09_30_2011.csv')
ttm_10 = pd.read_csv(ttm + 'ttm_09_30_2010.csv')
ttm_09 = pd.read_csv(ttm + 'ttm_09_30_2009.csv')

# importing sp500 constituents
sp500 = pd.read_csv(ttm + 'constituents_csv.csv')
```

```python
# concatenating all datasets into single dataframe
ttm_list = [ttm_18, ttm_17, ttm_16, ttm_15, ttm_14, ttm_13, ttm_12, ttm_11, ttm_10, ttm_09]
merged_financials = pd.concat(ttm_list, axis = 0, ignore_index= True, join = 'outer')
```

```python
merged_financials.head()
```

### Removing non voting shares from sp500 consituents
##### Also droping 'Financials' companies from dataset (out of scope for this project)
>symbols:
> GOOG,
> DISCK,
> NWS,
> FOX,
> UA.

```python
# Indexing the 5 non voting tickers from dataset and financial companies. 
indexNames = sp500[(sp500['ticker'] == 'GOOG') | (sp500['ticker'] == 'DISCK')
                     | (sp500['ticker'] == 'NWS') | (sp500['ticker'] == 'FOX')
                    | (sp500['ticker'] == 'UA') | (sp500['sector'] == 'financials')].index
print(indexNames)
```

```python
# Delete the row indexes above from dataFrame
sp500.drop(indexNames, inplace=True)
```

```python
# creating an list of the remaing 431 companies
symbols = sp500['ticker'].tolist()
print(type(symbols))
print(len(symbols))
```

```python
# filtering the dataframe with 
merged_financials = merged_financials[merged_financials['ticker'].isin(symbols)]
```

```python
# simplifying sector names
final_df = pd.merge(merged_financials, sp500[['ticker', 'sector']], left_on = ['ticker'],
                    right_on = ['ticker'], how = 'left')
```

```python
# dropping original sector name column
final_df.drop(['sector_x'], axis=1, inplace = True)
```

```python
# changing column orders
cols = final_df.columns.tolist()
cols = cols[-1:] + cols[:-1]
final_df = final_df[cols]
```

```python
# droping missing values
final_df.dropna(inplace = True)
```

```python
final_df
```

```python
# creating new dataset with companies that have at leats 3 years of financial data
companies = final_df.groupby('company')
companies = companies.filter(lambda x: len(x) > 3)
```

```python
# changing datatypes and using a dictionary to convert specific columns
companies.columns
```

```python
# removing '%' from ratio columns and changing it to float.
companies['gross_margin'] = companies['gross_margin'].str.rstrip('%').astype('float') / 100.0
companies['operating_margin'] = companies['operating_margin'].str.rstrip('%').astype('float') / 100.0
companies['net_margin'] = companies['net_margin'].str.rstrip('%').astype('float') / 100.0
companies['roe'] = companies['roe'].str.rstrip('%').astype('float') / 100.0
companies['roa'] = companies['roa'].str.rstrip('%').astype('float') / 100.0
companies['fcf_net_income'] = companies['fcf_net_income'].str.rstrip('%').astype('float') / 100.0
companies['current_ratio'] = companies['current_ratio'].str.rstrip('%').astype('float') / 100.0
companies['liab_equity'] = companies['liab_equity'].str.rstrip('%').astype('float') / 100.0
companies['debt_assets'] = companies['debt_assets'].str.rstrip('%').astype('float') / 100.0
```

```python
# Removing commans from all interger columns
cols = ['revenues', 'cogs', 'net_income', 'ebit', 'ebitda', 'chg_wc', 'r_and_d', 'assets', 'debt',
       'equity', 'dividends', 'cash_and_eq', 'curr_assets', 'curr_liab', 'sg_a','op_exp',
        'net_interest_paid', 'pretax_income', 'taxes']

# pass them to df.replace(), specifying each char and it's replacement:
companies[cols] = companies[cols].replace({',': ''}, regex=True)
```

```python
# converting string objects to integers
convert_dict = {'revenues':int,
'cogs':int,
'net_income':int,
'ebit':int,
'ebitda':int,
'chg_wc':int,
'r_and_d':int,
'assets':int,
'debt':int,
'equity':int,
'dividends':int,
'cash_and_eq':int,
'curr_assets':int,
'curr_liab':int,
'sg_a':int,
'op_exp':int,
'net_interest_paid':int,
'pretax_income':int,
'taxes':int}
companies = companies.astype(convert_dict)
```

```python
# converting period to date format
companies['period']= pd.to_datetime(companies['period'])
```

```python
companies.dtypes
```

#### Creating Financial Ratios additional financial ratios

```python
# payout_ratio
companies['pay_out_ratio'] = companies['dividends']*-1/companies['net_income']

# net_debt_ebitda
companies['net_debt_ebitda'] = (companies['debt'] - companies['cash_and_eq'])/companies['ebitda']

# net_interest_ebitda
companies['net_interest_ebitda'] = companies['net_interest_paid']/companies['ebitda']

# net_debt_equity
companies['net_debt_equity'] = (companies['debt'] - companies['cash_and_eq'])/companies['equity']

# asset_turn_over
companies['asset_turn_over'] = companies['revenues']/companies['assets']

# fcf
companies['free_cash_flow'] = round(companies['net_income'] * companies['fcf_net_income'],0).astype(int)

# fcf_dividends
companies['fcf_dividends'] = companies['dividends']*-1/companies['free_cash_flow']

# debt_to_fcf
companies['debt_to_fcf'] = (companies['debt'] - companies['cash_and_eq'])/companies['free_cash_flow']

# fcf_sales
companies['fcf_revenues'] = companies['free_cash_flow']/companies['revenues']

# interest_coverage_fcf
companies['interest_coverage_fcf'] = companies['net_interest_paid']/companies['free_cash_flow']

# sg_a_ratio
companies['sg_a_ratio'] = companies['sg_a']/companies['revenues']

# op_expense_ratio
companies['op_expense_ratio'] = companies['op_exp']/companies['revenues']

# r_and_d_ratio
companies['r_and_d_ratio'] = companies['r_and_d']/companies['revenues']

# cash_ratio
companies['cash_ratio'] = companies['cash_and_eq']/companies['curr_liab']

# wk_turover
companies['wk_turnover'] = (companies['curr_assets'] - companies['curr_liab'])/companies['revenues']

# ebitda_margin
companies['ebitda_margin'] = companies['ebitda']/companies['revenues']

# ebit_margin
companies['ebit_margin'] = companies['ebit']/companies['revenues']

# pre_tax_margin
companies['pre_tax_margin'] = companies['pretax_income']/companies['revenues']

#tax_to_revenues
companies['tax_to_revenues'] = companies['taxes']/companies['revenues']
```

```python
companies.columns
```

```python
# droping unecessary absolute numeric columns
companies.drop(cols, axis=1, inplace = True)
```

```python
# reseting the dataframe index
companies = companies.reset_index(drop=True)
```

```python
# creating a dataset with all tickers in order to get historical stocks' price changes on Yahoo
final_tickers = companies[['ticker', 'company', 'sector_y', 'period']].\
                sort_values(by = ['ticker', 'period']).reset_index(drop=True)
```

```python
# exporting both df to csv format.
companies.to_csv(ttm + 'final_companies_dataset.csv')
```
